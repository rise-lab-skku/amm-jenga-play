```python
import rospy
import tf
from geometry_msgs.msg import PointStamped
rospy.init_node('tf_study')
listener = tf.TransformListener()
broadcaster = tf.TransformBroadcaster()
r = rospy.Rate(.1)
```

```python
class test:
    def __init__(self) -> None:
        ps=PointStamped()
        ps.header.frame_id='test_link'
        self.ps=ps
        self.pub=rospy.Publisher('test_point',PointStamped,queue_size=1)
    
    def publish(self):
        self.ps.header.stamp=rospy.Time.now()
        self.pub.publish(self.ps)
```

```python
rt=test()
while not rospy.is_shutdown():
    rt.publish()
    r.sleep()
```
